```python
df = pd.DataFrame(np.random.random((5,5)), columns=["a","b","c","d","e"])
```

```python
def plot_corr(df):
    corr=df.corr()
    colormap = sns.diverging_palette(220, 10, as_cmap=True)
    
    sns.heatmap(corr, 
            xticklabels=corr.columns,
            yticklabels=corr.columns,cmap=colormap,annot=True,fmt=".2f")
```
